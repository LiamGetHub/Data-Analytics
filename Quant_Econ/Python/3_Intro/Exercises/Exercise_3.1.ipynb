```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = 200
a = 0.9 #correlation betwenn x_t and x{t+1}
x = [0]

for i in range(t):
    #new shock value each iteration
    e_value = np.random.randn()

    #add to list
    x.append( (x[i] * a) + e_value )

plt.plot(x, label='time series')
plt.legend()
plt.show()
```
