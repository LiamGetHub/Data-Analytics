```python
import numpy as np
import matplotlib.pyplot as plt

def abs_val(num):
    if num > 0:
        return num
    else:
        return (num * -1)
```

```python
#given solution
α_values = [0.0, 0.8, 0.98]
T = 200
x = np.empty(T+1)

for α in α_values:
    x[0] = 0
    for t in range(T):
        x[t+1] = α * abs_val(x[t]) + np.random.randn()
    plt.plot(x, label=f'$\\alpha = {α}$')

plt.legend()
plt.show()
```
