```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
α_scenarios = [0.0, 0.8, 0.98]
T = 200
x = np.empty(T+1)
x[0] = 0

for a in α_scenarios:
   x[0] = 0 # reset first element each time
   for t in range(T):
      x[t+1] = a * x[t] + np.random.randn()
   plt.plot(x)

plt.show()
```

```python
#given solution
α_values = [0.0, 0.8, 0.98]
T = 200
x = np.empty(T+1)

for α in α_values:
    x[0] = 0
    for t in range(T):
        x[t+1] = α * x[t] + np.random.randn()
    plt.plot(x, label=f'$\\alpha = {α}$')

plt.legend()
plt.show()
```
