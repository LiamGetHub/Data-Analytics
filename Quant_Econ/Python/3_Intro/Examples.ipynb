## An Introductory example


```python
import numpy as np
import matplotlib.pyplot as plt

_values = np.random.randn(100)
plt.plot(_values)
plt.show()
```

```python
#Same outcome as a code above but with a for loop
ts_length = 100
_values = []   # empty list

for i in range(ts_length):
    e = np.random.randn()
    _values.append(e)

plt.plot(_values)
plt.show()
```

```python
#example of for loop with use of lists. 
# Notice multiple data types. Notible methods: append, pop, push
#for variable_name in sequence:
#    <code block>

animals = ['dog', 'cat', 'bird']
for animal in animals:
    print("The plural of " + animal + " is " + animal + "s")

```

```python
#while loop example
ts_length = 100
_values = []
i = 0
while i < ts_length:
    e = np.random.randn()
    _values.append(e)
    i = i + 1
plt.plot(_values)
plt.show()
```

```python
#ploting value of bank acount over time given no withdrawls
r = 0.025         # interest rate
T = 50            # end date
b = np.empty(T+1) # an empty NumPy array, to store all b_t
b[0] = 10         # initial balance

for t in range(T):
    b[t+1] = (1 + r) * b[t]

plt.plot(b, label='bank balance')
plt.legend()
plt.show()
```
